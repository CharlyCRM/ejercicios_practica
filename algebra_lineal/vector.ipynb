```python
# Representación de un vector een su posición estandar (inicio del vector en las coordenadas 0,0)

import numpy as np
import matplotlib.pyplot as plt

# componentes del vecor v = 2, -1
vx, vy = 2, -1 

plt.quiver(
    0, 0,      # Coordenadas de origen
    vx, vy,    # componentes del vector
    angles='xy',
    scale_units='xy',
    scale=1,
    color='red'
)

plt.axhline(0, color='black')  # eje X
plt.axvline(0, color='black')  # eje Y
plt.grid(True, linestyle='--', alpha=0.6)
plt.axis([-3, 3, -3, 3])
plt.gca().set_aspect('equal', adjustable='box') # Esto evita distorsiones geométricas.


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([1, 2]) # vector

for _ in range(10):
    s = np.random.randn() # escalar
    sv = s * v

    color = 'red' if s >= 0 else 'blue'

    plt.quiver(
        0, 0, # origen
        sv[0], sv[1], # componentes del vector
        angles='xy',
        scale_units='xy',
        scale=1,
        color=color,
        alpha=0.7
    )

plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axis([-5, 5, -5, 5])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
