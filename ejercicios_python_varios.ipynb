## Enunciado del Hito 1

Dispones de una lista con los importes de varias ventas (números).

Ejemplo conceptual (no código):

	• venta 1 → 12.50 €
	• venta 2 → 7.80 €
	• venta 3 → 21.00 €
	• venta 4 → 5.40 €

Queremos analizar esta lista para obtener información básica:

	1. El total de ingresos
	2. El número total de ventas
	3. Cuántas ventas son mayores de 10 €

```python
lista_ingresos: list = [12.40, 7.80, 21.00, 5.40]
margen_venta = 7.00

def total_ingresos(lista: list) -> float:
    """
    Calcula el total de ingresos
    """
    total_ingresos = sum(lista)
    return total_ingresos

def numero_total_ventas(lista: list) -> int:
    """
    Calcula el número total de ventas
    """
    numero_total_ventas = len(lista)
    return numero_total_ventas

def ventas_mayores_x(lista: list, x:float = 0) -> int:
    venta_mayores = 0
    for i in lista:
        if i >= x:
            venta_mayores += 1
    return venta_mayores
        
# Pruebas
print(f"Total ingresos: {total_ingresos(lista_ingresos)}€")
print(f"Número total de ventas: {numero_total_ventas(lista_ingresos)}")
print(f"Venta mayores a {margen_venta}: {ventas_mayores_x(lista_ingresos, margen_venta)}")
```

## Enunciado del Hito 2

Tienes una lista de ventas.

	• producto
	• cantidad
	• precio_unitario

Cada venta será un diccionario.
Cada venta es un diccionario con la información anterior.

Queremos calcular:

	• el importe total de cada venta (cantidad × precio_unitario)
	• el total de ingresos de todas las ventas

```python
ventas = [
    {"producto": "manzana", "cantidad": 3, "precio_unitario": 0.5},
    {"producto": "pera", "cantidad": 5, "precio_unitario": 0.4},
    {"producto": "manzana", "cantidad": 2, "precio_unitario": 0.5},
    {"producto": "plátano", "cantidad": 4, "precio_unitario": 0.3},
]

def importe_total(datos: list) -> list:
    """
    Calcula el importe total de cada ventas
    """
    importe_total = []
    
    for i in datos:
        importe_venta = i["cantidad"] * i["precio_unitario"]
        importe_total.append(importe_venta)
    return importe_total

def total_ingresos(datos: list) -> float:
    """
    Calcula el total de ingresos
    """
    total_ingresos = 0

    for i in datos:
        importe_ingresos = i["cantidad"] * i["precio_unitario"]
        total_ingresos += importe_ingresos
    return total_ingresos

#Pruebas
print(f"Importe total de cada venta: {importe_total(ventas)}")
print(f"Total de ingresos: {total_ingresos(ventas)}€")
```
