### Reto #0 ‚Äî El famoso FizzBuzz

En qu√© consiste

## Reto #1 ‚Äî ¬øEs un anagrama?

En qu√© consiste

El reto plantea crear una funci√≥n que reciba dos palabras y determine si una es un anagrama de la otra.

Dos palabras son anagramas si:
- Contienen las mismas letras
- En la misma cantidad
- En distinto orden (o no necesariamente en el mismo)

Ejemplos conceptuales:

- "roma" y "amor" ->
- "listen" y "silent"
- "hola" y "halo"

```python
def es_anagrama(palabra1: str, palabra2: str) -> bool:
    palabra1 = palabra1.lower()
    palabra2 = palabra2.lower()

    # Si son exactamente iguales (tras normalizar), no se considera anagrama
    if palabra1 == palabra2:
        return False

    # Si tienen distinta longitud, imposible que sean anagramas
    if len(palabra1) != len(palabra2):
        return False

    # Si al ordenar letras quedan iguales, son anagramas
    return sorted(palabra1) == sorted(palabra2)

#-----------------------------
print(es_anagrama("ROMA", "amor"))  # True
print(es_anagrama("hola", "hola"))  # False
print(es_anagrama("aab", "abb"))    # False
```

## Reto #2 ‚Äî ¬øEs un pal√≠ndromo?

En qu√© consiste

Debes crear una funci√≥n que determine si un texto es un pal√≠ndromo.

Un pal√≠ndromo es una palabra o frase que se lee igual de izquierda a derecha que de derecha a izquierda, ignorando:
- may√∫sculas y min√∫sculas
- espacios

Ejemplos conceptuales:
- "radar" ‚Üí s√≠
- "Ana" ‚Üí s√≠
- "Anita lava la tina" ‚Üí s√≠
- "Hola mundo" ‚Üí no

```python
def es_palindromo(palabra: str) -> bool:
    """
    Indica si una palabra es pal√≠ndromo
    """
    # Minuscula + eliminaci√≥n de esp√°cios
    palabra = palabra.lower().replace(" ","")
    palabra_reves = palabra[::-1]

    return palabra == palabra_reves

#--------------------------------
print(es_palindromo("radar"))
print(es_palindromo("Ana"))
print(es_palindromo("Anita lava la TINA"))
print(es_palindromo("Hola Mundo"))
```

En qu√© consiste

Debes crear una funci√≥n que determine si un n√∫mero entero es primo.

Un n√∫mero primo es aquel que:
- es mayor que 1
- solo es divisible por 1 y por s√≠ mismo

Ejemplos:
- 2 ‚Üí primo
- 3 ‚Üí primo
- 4 ‚Üí no primo (2 √ó 2)
- 7 ‚Üí primo
- 1 ‚Üí no primo
- 0 y n√∫meros negativos ‚Üí no primos

```python
"""
Los n√∫meros primos son n√∫meros naturales mayores que 1 que solo tienen dos divisores exactos: el 1 y ellos mismos, 
es decir, no se pueden dividir por ning√∫n otro n√∫mero entero sin dejar residuo. Ejemplos comunes son 2, 3, 5, 7, 11, etc., 
mientras que los n√∫meros con m√°s de dos divisores se llaman compuestos (como el 6, que es divisible por 1, 2, 3 y 6). 

El n√∫mero 1 no es primo ni compuesto, ya que solo tiene un divisor.  
"""

import math

def es_primo(n: int) -> bool:
    """
    Determina si un n√∫mero es primo
    """
    if n <= 1:
        return False
    # n ** 0,5 = n elevado a 0,5
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


print(es_primo(2)) # primo
print(es_primo(3)) # primo
print(es_primo(4)) # no primo
print(es_primo(7)) # primo
print(es_primo(1)) # no primo
print("*--------------------------")
print(f"9 elevado a 0,5: {9**0.5}")
print(f"Raiz cuadrada de 9: {math.sqrt(9)}")

"""
Para determinar si un n√∫mero es primo, no es necesario comprobar todos los posibles divisores hasta el propio n√∫mero. 
Si un n√∫mero n no es primo, entonces puede expresarse como el producto de dos enteros a ¬∑ b = n, ambos mayores que 1. 

En ese caso, al menos uno de esos factores debe ser menor o igual que la ra√≠z cuadrada de n. 
Si ambos fueran mayores que ‚àön, su producto ser√≠a mayor que n, lo cual es imposible. 

Por este motivo, basta con comprobar si existe alg√∫n divisor entre 2 y ‚àön. 
Si no se encuentra ninguno en ese rango, el n√∫mero no puede tener divisores mayores y, por tanto, es primo.

    29
    ‚Ä¢	‚àö29 ‚âà 5.38
    ‚Ä¢	pruebo divisores: 2, 3, 4, 5
    ‚Ä¢	ninguno divide ‚Üí primo

    49
	‚Ä¢	‚àö49 = 7
	‚Ä¢	pruebo 2, 3, 4, 5, 6, 7
	‚Ä¢	7 divide ‚Üí no primo
"""
None
```

## Reto #4 ‚Äî Invertir una cadena

En qu√© consiste

Debes crear una funci√≥n que reciba un texto y devuelva ese mismo texto invertido.

Ejemplos:
- "Hola" ‚Üí "aloH"

‚ö†Ô∏è No hay que eliminar espacios ni normalizar nada (eso ya lo hicimos en otros retos).

```python
def invertir_cadena(cadena: str) -> str:
    """
    Recibe un texto y devuelve el mismo texto invertido
    """
    return cadena[::-1]


print(invertir_cadena("Hola"))
print(invertir_cadena("Python"))
print(invertir_cadena("12345"))
```

## Reto #5 ‚Äî Aspect Ratio de una imagen

En qu√© consiste

Debes crear una funci√≥n que reciba el ancho y el alto de una imagen (dos enteros positivos) y devuelva su aspect ratio en formato: 'ancho:alto'

Ejemplos conceptuales
- (1920, 1080) ‚Üí "16:9"
- (1280, 720) ‚Üí "16:9"
- (1024, 768) ‚Üí "4:3"
- (100, 100) ‚Üí "1:1"

Pistas para enfocar el problema (sin c√≥digo)

	1.	Qu√© significa ‚Äúsimplificar‚Äù
No es dividir a ojo. Es reducir ambos n√∫meros por el mismo factor hasta que no se pueda m√°s.

	2.	Idea matem√°tica clave
Si divides ancho y alto por su m√°ximo com√∫n divisor (MCD), obtienes la forma simplificada.

	3.	Flujo l√≥gico
	‚Ä¢	Recibes ancho y alto
	‚Ä¢	Calculas un valor com√∫n que los divide a ambos
	‚Ä¢	Divides ambos por ese valor
	‚Ä¢	Devuelves el resultado como texto "a:b"
	4.	Restricciones
	‚Ä¢	Asume n√∫meros enteros positivos
	‚Ä¢	El resultado debe ser un str, no una tupla

```python
def calcular_aspect_ratio(ancho_original: int, alto_original: int) -> str:
    """
    Devuelve el aspect ratio de una imagen en formato 'ancho:alto' simplificado.
    """
    a = ancho_original
    b = alto_original

    # Algoritmo de Euclides para el MCD
    while b != 0:
        a, b = b, a % b

    mcd = a
    ancho_simplificado = ancho_original // mcd
    alto_simplificado = alto_original // mcd

    return f"{ancho_simplificado}:{alto_simplificado}"


print(calcular_aspect_ratio(48, 18))     # 8:3
print(calcular_aspect_ratio(1920, 1080)) # 16:9
print(calcular_aspect_ratio(1024, 768))  # 4:3
```

### üß© Mini-reto 2 ‚Äî Calcular solo el MCD

Reglas:

	‚Ä¢	Recibe dos n√∫meros a y b
	‚Ä¢	Usa while
	‚Ä¢	Usa a, b = b, a % b
	‚Ä¢	Devuelve un solo n√∫mero: el MCD
	‚Ä¢	Nada de strings, nada de ratios, nada m√°s

Piensa en una divisi√≥n normal. Si divides 17 entre 5, sabes que 5 cabe 3 veces y sobran 2. Ese ‚Äú2‚Äù es el resto. En matem√°ticas se suele escribir:

17 = 3¬∑5 + 2

Aqu√≠, 2 es el resto. A ese resto se le suele llamar r solo como nombre, igual que podr√≠as llamarlo ‚Äúresto‚Äù, ‚Äúsobrante‚Äù o ‚Äúx‚Äù.

Cuando en programaci√≥n escribes 17 % 5, Python hace exactamente esa cuenta y te devuelve el resto. Es decir:

17 % 5 ‚Üí 2

El s√≠mbolo % se suele llamar ‚Äúmod‚Äù. As√≠ que mod no es un n√∫mero, es la operaci√≥n que calcula el resto.

Por eso aparecen las dos cosas a la vez. Primero se define la idea general:

a = q¬∑b + r

Luego se dice c√≥mo se calcula ese r en la pr√°ctica:

r = a % b

```python
# MCD(a, b) = MCD(b, a % b) -> Algoritmo de Euclides para hayar el MCD
a = 48
b = 18

while b != 0:
    a, b = b, a % b

print(f"E MCD es: {a}")
print(f"El resto es: {b}")
```

### üß© Mini-reto 3 ‚Äî COPRIMOS

Escribe una funci√≥n que reciba dos n√∫meros enteros positivos y devuelva True si son coprimos y False en caso contrario.

Qu√© significa ‚Äúcoprimos‚Äù

Dos n√∫meros son coprimos si su MCD es 1.

Ejemplos:

	‚Ä¢	8 y 15 ‚Üí coprimos ‚Üí True
	‚Ä¢	12 y 18 ‚Üí no coprimos ‚Üí False
	‚Ä¢	14 y 25 ‚Üí coprimos ‚Üí True

```python
def es_copirmo(a: int, b: int) -> bool:
    """
    Calcula si dos n√∫meros son coprimos
    """
    while b != 0:
        a, b = b, a % b # Calcula el MCD con el algoritmo de euclides
    
    if a == 1: # comprueba si es copirmo
        return True
    else:
        return False
    
#---------------------------------
print(es_copirmo(8,15))
print(es_copirmo(12,18))
print(es_copirmo(14,25))
```

## Reto #6 ‚Äî Contar Vocales en un texto

En qu√© consiste

Escribir una funci√≥n que reciba un texto y devuelva cu√°ntas vocales contiene.

Ejemplos:

    ‚Ä¢   "Hola" --> 2
    ‚Ä¢   "Python --> 1
    ‚Ä¢   "aeiou" --> 5
    ‚Ä¢   "BCDF" --> 0

```python
def contar_vocales(texto: str) -> int:
    """
    Cuenta el n√∫mero de vocales de un texto
    """
    VOCALES = ("a", "e", "i", "o", "u")
    contador = 0

    for letra in texto.lower():
        if letra in VOCALES:
            contador += 1
    return contador

#--------------------------------------
print(contar_vocales("Hola"))
print(contar_vocales("Python"))
print(contar_vocales("aeiou"))
print(contar_vocales("BCDF"))
```

## Reto #4.1 ‚Äî √ÅREA DE UN POL√çGONO

Debes crear una funci√≥n que calcule el √°rea de un pol√≠gono regular dados:

	‚Ä¢	el n√∫mero de lados
	‚Ä¢	la longitud de cada lado

Casos t√≠picos:

	‚Ä¢	tri√°ngulo
	‚Ä¢	cuadrado
	‚Ä¢	pent√°gono
	‚Ä¢	etc.

```python
import math

def area_poligono_regular(lados: int, longitud: float) -> float:
    """
    Calcula el area de un poligono regular
    """
    # Area = (perimetro x Apotema) / 2
    # Perimetro = lados * longitud
    # apotema = (longitud / 2) / tan(œÄ / n)
    # n = lados

    perimetro = lados * longitud
    apotema = (longitud / 2) / math.tan(math.pi / lados)

    return round((perimetro * apotema) / 2, 2)

#----------------------
print(area_poligono_regular(4, 2.4))
print(area_poligono_regular(3, 5.8))

```

## Reto #8 ‚Äî DECIMAL A BINARIO

Debes crear una funci√≥n que reciba un n√∫mero entero positivo en base 10 y devuelva su representaci√≥n en binario (base 2), sin usar funciones integradas que ya lo hagan por ti.

Ejemplos:

	‚Ä¢	5 ‚Üí "101"
	‚Ä¢	10 ‚Üí "1010"
	‚Ä¢	1 ‚Üí "1"
	‚Ä¢	0 ‚Üí "0" (decide t√∫ c√≥mo manejar este caso, pero hazlo expl√≠cito)

```python
def generar_binario(numero: int) -> str:
    """
    Transforma un numero en base 10 en su correspondiente binario
    """
    contador =""
    while numero > 0:
        resto, numero = numero % 2, numero // 2

        if resto == 0 or resto == 1:
            contador += str(resto)
    return contador[::-1]

#------------------------------------------
print(generar_binario(5))
print(generar_binario(10))
print(generar_binario(1))
```
