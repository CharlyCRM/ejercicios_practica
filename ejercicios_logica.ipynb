### Reto #0 — El famoso FizzBuzz

En qué consiste

El primer reto es un clásico de programación y lógica.

Debes escribir un programa que recorra los números del 1 al 100 y, para cada número:
- Muestre “Fizz” si el número es múltiplo de 3
- Muestre “Buzz” si el número es múltiplo de 5
- Muestre “FizzBuzz” si es múltiplo de 3 y 5 a la vez
- En cualquier otro caso, muestre el número tal cual

El objetivo no es la dificultad técnica, sino entrenar:
- Condicionales
- Orden lógico de las comprobaciones
- Lectura clara del código

```python
# Genero un lista del 1 al 100
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 5 == 0:
        print("Buzz")
    elif i % 3 == 0:
        print("Fizz")
    else:
        print(i)
```

## Reto #1 — ¿Es un anagrama?

En qué consiste

El reto plantea crear una función que reciba dos palabras y determine si una es un anagrama de la otra.

Dos palabras son anagramas si:
- Contienen las mismas letras
- En la misma cantidad
- En distinto orden (o no necesariamente en el mismo)

Ejemplos conceptuales:

- "roma" y "amor" ->
- "listen" y "silent"
- "hola" y "halo"

```python
def es_anagrama(palabra1: str, palabra2: str) -> bool:
    palabra1 = palabra1.lower()
    palabra2 = palabra2.lower()

    # Si son exactamente iguales (tras normalizar), no se considera anagrama
    if palabra1 == palabra2:
        return False

    # Si tienen distinta longitud, imposible que sean anagramas
    if len(palabra1) != len(palabra2):
        return False

    # Si al ordenar letras quedan iguales, son anagramas
    return sorted(palabra1) == sorted(palabra2)

#-----------------------------
print(es_anagrama("ROMA", "amor"))  # True
print(es_anagrama("hola", "hola"))  # False
print(es_anagrama("aab", "abb"))    # False
```

## Reto #2 — ¿Es un palíndromo?

En qué consiste

Debes crear una función que determine si un texto es un palíndromo.

Un palíndromo es una palabra o frase que se lee igual de izquierda a derecha que de derecha a izquierda, ignorando:
- mayúsculas y minúsculas
- espacios

Ejemplos conceptuales:
- "radar" → sí
- "Ana" → sí
- "Anita lava la tina" → sí
- "Hola mundo" → no

```python
def es_palindromo(palabra: str) -> bool:
    """
    Indica si una palabra es palíndromo
    """
    # Minuscula + eliminación de espácios
    palabra = palabra.lower().replace(" ","")
    palabra_reves = palabra[::-1]

    return palabra == palabra_reves

#--------------------------------
print(es_palindromo("radar"))
print(es_palindromo("Ana"))
print(es_palindromo("Anita lava la TINA"))
print(es_palindromo("Hola Mundo"))
```

## Reto #3 — ¿Es un número primo?

En qué consiste

Debes crear una función que determine si un número entero es primo.

Un número primo es aquel que:
- es mayor que 1
- solo es divisible por 1 y por sí mismo

Ejemplos:
- 2 → primo
- 3 → primo
- 4 → no primo (2 × 2)
- 7 → primo
- 1 → no primo
- 0 y números negativos → no primos

```python
"""
Los números primos son números naturales mayores que 1 que solo tienen dos divisores exactos: el 1 y ellos mismos, 
es decir, no se pueden dividir por ningún otro número entero sin dejar residuo. Ejemplos comunes son 2, 3, 5, 7, 11, etc., 
mientras que los números con más de dos divisores se llaman compuestos (como el 6, que es divisible por 1, 2, 3 y 6). 

El número 1 no es primo ni compuesto, ya que solo tiene un divisor.  
"""

import math

def es_primo(n: int) -> bool:
    """
    Determina si un número es primo
    """
    if n <= 1:
        return False
    # n ** 0,5 = n elevado a 0,5
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


print(es_primo(2)) # primo
print(es_primo(3)) # primo
print(es_primo(4)) # no primo
print(es_primo(7)) # primo
print(es_primo(1)) # no primo
print("*--------------------------")
print(f"9 elevado a 0,5: {9**0.5}")
print(f"Raiz cuadrada de 9: {math.sqrt(9)}")

"""
Para determinar si un número es primo, no es necesario comprobar todos los posibles divisores hasta el propio número. 
Si un número n no es primo, entonces puede expresarse como el producto de dos enteros a · b = n, ambos mayores que 1. 

En ese caso, al menos uno de esos factores debe ser menor o igual que la raíz cuadrada de n. 
Si ambos fueran mayores que √n, su producto sería mayor que n, lo cual es imposible. 

Por este motivo, basta con comprobar si existe algún divisor entre 2 y √n. 
Si no se encuentra ninguno en ese rango, el número no puede tener divisores mayores y, por tanto, es primo.

    29
    •	√29 ≈ 5.38
    •	pruebo divisores: 2, 3, 4, 5
    •	ninguno divide → primo

    49
	•	√49 = 7
	•	pruebo 2, 3, 4, 5, 6, 7
	•	7 divide → no primo
"""
None
```
