## Reto #1 â€” Â¿Es un anagrama?

En quÃ© consiste

El reto plantea crear una funciÃ³n que reciba dos palabras y determine si una es un anagrama de la otra.

Dos palabras son anagramas si:
- Contienen las mismas letras
- En la misma cantidad
- En distinto orden (o no necesariamente en el mismo)

Ejemplos conceptuales:

- "roma" y "amor" ->
- "listen" y "silent"
- "hola" y "halo"

```python
def es_anagrama(palabra1: str, palabra2: str) -> bool:
    palabra1 = palabra1.lower()
    palabra2 = palabra2.lower()

    # Si son exactamente iguales (tras normalizar), no se considera anagrama
    if palabra1 == palabra2:
        return False

    # Si tienen distinta longitud, imposible que sean anagramas
    if len(palabra1) != len(palabra2):
        return False

    # Si al ordenar letras quedan iguales, son anagramas
    return sorted(palabra1) == sorted(palabra2)

#-----------------------------
print(es_anagrama("ROMA", "amor"))  # True
print(es_anagrama("hola", "hola"))  # False
print(es_anagrama("aab", "abb"))    # False
```

## Reto #2 â€” Â¿Es un palÃ­ndromo?

En quÃ© consiste

Debes crear una funciÃ³n que determine si un texto es un palÃ­ndromo.

Un palÃ­ndromo es una palabra o frase que se lee igual de izquierda a derecha que de derecha a izquierda, ignorando:
- mayÃºsculas y minÃºsculas

Ejemplos conceptuales:
- "radar" â†’ sÃ­

```python
def es_palindromo(palabra: str) -> bool:
    """
    Indica si una palabra es palÃ­ndromo
    """
    # Minuscula + eliminaciÃ³n de espÃ¡cios
    palabra = palabra.lower().replace(" ","")
    palabra_reves = palabra[::-1]

    return palabra == palabra_reves

#--------------------------------
print(es_palindromo("radar"))
print(es_palindromo("Ana"))
print(es_palindromo("Anita lava la TINA"))
print(es_palindromo("Hola Mundo"))
```

## Reto #5 â€” Aspect Ratio de una imagen

En quÃ© consiste

Debes crear una funciÃ³n que reciba el ancho y el alto de una imagen (dos enteros positivos) y devuelva su aspect ratio en formato: 'ancho:alto'

Ejemplos conceptuales
- (1920, 1080) â†’ "16:9"
- (1280, 720) â†’ "16:9"
- (1024, 768) â†’ "4:3"
- (100, 100) â†’ "1:1"

Pistas para enfocar el problema (sin cÃ³digo)

	1.	QuÃ© significa â€œsimplificarâ€
No es dividir a ojo. Es reducir ambos nÃºmeros por el mismo factor hasta que no se pueda mÃ¡s.

	2.	Idea matemÃ¡tica clave
Si divides ancho y alto por su mÃ¡ximo comÃºn divisor (MCD), obtienes la forma simplificada.

	3.	Flujo lÃ³gico
	â€¢	Recibes ancho y alto
	â€¢	Calculas un valor comÃºn que los divide a ambos
	â€¢	Divides ambos por ese valor
	â€¢	Devuelves el resultado como texto "a:b"
	4.	Restricciones
	â€¢	Asume nÃºmeros enteros positivos
	â€¢	El resultado debe ser un str, no una tupla

```python
def calcular_aspect_ratio(ancho_original: int, alto_original: int) -> str:
    """
    Devuelve el aspect ratio de una imagen en formato 'ancho:alto' simplificado.
    """
    a = ancho_original
    b = alto_original

    # Algoritmo de Euclides para el MCD
    while b != 0:
        a, b = b, a % b

    mcd = a
    ancho_simplificado = ancho_original // mcd
    alto_simplificado = alto_original // mcd

    return f"{ancho_simplificado}:{alto_simplificado}"


print(calcular_aspect_ratio(48, 18))     # 8:3
print(calcular_aspect_ratio(1920, 1080)) # 16:9
print(calcular_aspect_ratio(1024, 768))  # 4:3
```

### ðŸ§© Mini-reto 2 â€” Calcular solo el MCD

Reglas:

	â€¢	Recibe dos nÃºmeros a y b

AquÃ­, 2 es el resto. A ese resto se le suele llamar r solo como nombre, igual que podrÃ­as llamarlo â€œrestoâ€, â€œsobranteâ€ o â€œxâ€.

Cuando en programaciÃ³n escribes 17 % 5, Python hace exactamente esa cuenta y te devuelve el resto. Es decir:

El sÃ­mbolo % se suele llamar â€œmodâ€. AsÃ­ que mod no es un nÃºmero, es la operaciÃ³n que calcula el resto.

Por eso aparecen las dos cosas a la vez. Primero se define la idea general:

a = qÂ·b + r

Luego se dice cÃ³mo se calcula ese r en la prÃ¡ctica:

r = a % b

```python
# MCD(a, b) = MCD(b, a % b) -> Algoritmo de Euclides para hayar el MCD
a = 48
b = 18

while b != 0:
    a, b = b, a % b

print(f"E MCD es: {a}")
print(f"El resto es: {b}")
```

### ðŸ§© Mini-reto 3 â€” COPRIMOS

Escribe una funciÃ³n que reciba dos nÃºmeros enteros positivos y devuelva True si son coprimos y False en caso contrario.

QuÃ© significa â€œcoprimosâ€

Dos nÃºmeros son coprimos si su MCD es 1.

Ejemplos:

	â€¢	8 y 15 â†’ coprimos â†’ True
	â€¢	12 y 18 â†’ no coprimos â†’ False
	â€¢	14 y 25 â†’ coprimos â†’ True

```python
def es_copirmo(a: int, b: int) -> bool:
    """
    Calcula si dos nÃºmeros son coprimos
    """
    while b != 0:
        a, b = b, a % b # Calcula el MCD con el algoritmo de euclides
    
    if a == 1: # comprueba si es copirmo
        return True
    else:
        return False
    
#---------------------------------
print(es_copirmo(8,15))
print(es_copirmo(12,18))
print(es_copirmo(14,25))
```

## Reto #6 â€” Contar Vocales en un texto

Escribir una funciÃ³n que reciba un texto y devuelva cuÃ¡ntas vocales contiene.

Ejemplos:

    â€¢   "Hola" --> 2
    â€¢   "Python --> 1
    â€¢   "aeiou" --> 5
    â€¢   "BCDF" --> 0

```python
def contar_vocales(texto: str) -> int:
    """
    Cuenta el nÃºmero de vocales de un texto
    """
    VOCALES = ("a", "e", "i", "o", "u")
    contador = 0

    for letra in texto.lower():
        if letra in VOCALES:
            contador += 1
    return contador

#--------------------------------------
print(contar_vocales("Hola"))
print(contar_vocales("Python"))
print(contar_vocales("aeiou"))
print(contar_vocales("BCDF"))
```

## Reto #4.1 â€” ÃREA DE UN POLÃGONO

Debes crear una funciÃ³n que calcule el Ã¡rea de un polÃ­gono regular dados:

	â€¢	el nÃºmero de lados
	â€¢	la longitud de cada lado

Casos tÃ­picos:

	â€¢	triÃ¡ngulo
	â€¢	cuadrado
	â€¢	pentÃ¡gono
	â€¢	etc.

```python
import math

def area_poligono_regular(lados: int, longitud: float) -> float:
    """
    Calcula el area de un poligono regular
    """
    # Area = (perimetro x Apotema) / 2
    # Perimetro = lados * longitud
    # apotema = (longitud / 2) / tan(Ï€ / n)
    # n = lados

    perimetro = lados * longitud
    apotema = (longitud / 2) / math.tan(math.pi / lados)

    return round((perimetro * apotema) / 2, 2)

#----------------------
print(area_poligono_regular(4, 2.4))
print(area_poligono_regular(3, 5.8))

```

## Reto 9 â€” Serie de Fibonacci

En quÃ© consiste

Debes crear una funciÃ³n que reciba un nÃºmero entero n y devuelva los n primeros nÃºmeros de la serie de Fibonacci.

La serie de Fibonacci se define asÃ­:

	â€¢	Los dos primeros valores son 0 y 1
	â€¢	Cada nÃºmero siguiente es la suma de los dos anteriores

Ejemplos:

	â€¢	n = 1 â†’ [0]
	â€¢	n = 2 â†’ [0, 1]
	â€¢	n = 5 â†’ [0, 1, 1, 2, 3]
	â€¢	n = 8 â†’ [0, 1, 1, 2, 3, 5, 8, 13]

```python
def serie_fibonacci(n: int) -> list:
    """
    Genera los n primeros nÃºmeros de la serie Fibonacci
    """
    if n == 0:
        return []
    if n == 1:
        return [0]
    
    fibonacci = [0, 1]
    previo, actual = 0, 1

    for _ in range(n - 2):
        previo, actual = actual, previo + actual
        fibonacci.append(actual)
    return fibonacci

#------------------------------------
print(serie_fibonacci(1))
print(serie_fibonacci(2))
print(serie_fibonacci(5))
print(serie_fibonacci(8))
```

## Reto 10 â€” Factorial de un nÃºmero

Crea una funciÃ³n que reciba un entero no negativo n y devuelva su factorial.

Por definiciÃ³n:

	â€¢	0! = 1
	â€¢	1! = 1
	â€¢	n! = n Ã— (n-1) Ã— (n-2) Ã— â€¦ Ã— 1

Ejemplos:

	â€¢	factorial(0) â†’ 1
	â€¢	factorial(1) â†’ 1
	â€¢	factorial(5) â†’ 120
	â€¢	factorial(7) â†’ 5040


```python
def factorial(n: int) -> int:
    """
    Devuelve el factorial de un nÃºmero
    """
    if n == 0:
        return 1
    if n == 1:
        return 1
    
    acumulador = 1
    contador = 2 # 2 hasta n

    while contador != n + 1:
        resultado = acumulador * contador
        contador += 1
        acumulador = resultado
    return acumulador

#----------------------------
print(factorial(0))
print(factorial(1))
print(factorial(5))
print(factorial(7))
```

```python
def factorial(n: int) -> int:
    if n < 0:
        raise ValueError("El factorial no estÃ¡ definido para nÃºmeros negativos")

    resultado = 1
    for i in range(2, n + 1):
        resultado *= i
    return resultado

#----------------------------
print(factorial(0))
print(factorial(1))
print(factorial(5))
print(factorial(7))
```

## Reto 11 â€” Â¿Es un nÃºmero perfecto?

Crea una funciÃ³n que reciba un nÃºmero entero positivo y determine si es un nÃºmero perfecto.

Un nÃºmero perfecto es aquel que es igual a la suma de sus divisores propios (divisores positivos menores que Ã©l, excluyÃ©ndose a sÃ­ mismo).

Ejemplos

	â€¢	6 â†’ Perfecto
Divisores: 1, 2, 3 â†’ 1 + 2 + 3 = 6

	â€¢	28 â†’ Perfecto
Divisores: 1, 2, 4, 7, 14 â†’ suma 28

	â€¢	12 â†’ No perfecto
	â€¢	1 â†’ No perfecto


```python
def es_perfecto(n: int) -> bool:
    """
    Indica si un nÃºmero es perfecto
    """
    if n <= 1:
        raise ValueError("Solo se adminten valores mayores a 1")
    
    if n > 1:
        divisores = set() # almaceno los divisores propios
        for i in range (1, n):
            if n % i == 0:
                divisores.add(i)
        suma = sum(divisores)
        return suma == n

#-------------------
print(es_perfecto(6))
print(es_perfecto(28))
print(es_perfecto(12))
```
