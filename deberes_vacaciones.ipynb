```python
# Opcion 1
ciudades = [" Barcelona ", "madrid", "Valencia", "madrid", " barcelona", "SEVILLA ", "valencia "]
lista_limpia = []
# Eliminar espacios en blanco
for ciudad in ciudades:
    ciudad_limpia = ciudad.strip()
    lista_limpia.append(ciudad_limpia)

print(f"Lista con los valores sin espacios: {lista_limpia}")

# -----------------------------

# Nombres en min√∫sculas
lista_minusculas = []
for i in lista_limpia:
    minusculas = i.lower()
    lista_minusculas.append(minusculas)

print(f"\nLista en min√∫sculas: {lista_minusculas}")

# -------------------------

# Ordenar Alfabeticamente
print(f"\nLista ordenada: {sorted(lista_minusculas)}")

# Elminar duplicados
sin_duplicados = list(set(lista_minusculas))
print(f"\nLista sin duplicados: {sin_duplicados}")

# lista final
lista_final = sorted(list(set(lista_minusculas)))
print(f"\nLista limpia, sin duplicados y ordenada: {lista_final}")

# ------------------------

# Opcion 2 
ciudades = [" Barcelona ", "madrid", "Valencia", "madrid", " barcelona", "SEVILLA ", "valencia "]

# Paso 1: limpiar espacios y pasar a min√∫sculas en la misma lista
for i in range(len(ciudades)):
    ciudades[i] = ciudades[i].strip().lower()

# Paso 2: eliminar duplicados y ordenar
ciudades = sorted(list(set(ciudades)))

print(ciudades)
```

```python
nombres = [
    "Carlos", "  ", "lucia", "m4rio", "Ana", "carlos", "ana", "", "Lucia", "   ",
    "juan", "Ju4n", "MARTA", "Marta", "sofia", "sof√≠a"
]

lista_limpia = []

for nombre in nombres:
    nombre = nombre.strip().lower()
    contiene_numero = False
    
    for letra in nombre:
        if letra.isdigit():
            contiene_numero = True
    
    if nombre and not contiene_numero:
        lista_limpia.append(nombre)

lista_limpia = sorted(list(set(lista_limpia)))
print(f"Lista limpia: {lista_limpia}")
```

```python
productos = ["pan", "leche", "huevos", "galletas"]

# Mostrar cada producto en may√∫sculas
for producto in productos:
    producto = producto.upper()
    print(f"Nombre del producto en Mayuscula: {producto}")
print(f"Lista original {productos}")

# üîπ Solo lectura y visualizaci√≥n
# 	‚Ä¢	Aqu√≠ est√°s recorriendo la lista original con una variable temporal producto.
# 	‚Ä¢	Le aplicas .upper() solo a esa variable, pero no se modifica la lista original.
# 	‚Ä¢	Imprimes el valor modificado, pero cuando el bucle acaba, la lista sigue igual.
# üß† Met√°fora: es como si ves un objeto, le pones unas gafas para verlo diferente, pero luego no lo tocas.

print("="*30)

# Reescribir la lista convirtiendo cada producto a may√∫sculas

for index in range(len(productos)):
    productos[index] = productos[index].upper()

print(f"Nombre del producto en Mayuscula: {productos}")

# üîπ Modifica la lista Original

print("="*30)

# Mostrar cada producto con su n√∫mero de posici√≥n (empezando en 1)
for index, valor in enumerate(productos, start=1):
    print(f"{index}. {valor}")
```

```python
cesta = [
    ("pan", 1.1),
    ("leche", 0.9),
    ("huevos", 2.5),
    ("galletas", 1.8),
    ("leche", 0.9),
    ("pan", 1.1)
]

# Crea una lista sin productos repetidos (por nombre), manteniendo solo uno.
cesta_unica = []
productos_ya_vistos = set()

for producto, precio in cesta:
    if producto not in productos_ya_vistos:
        productos_ya_vistos.add(producto)
        cesta_unica.append((producto, precio))

print(cesta_unica)
```

```python
ventas = {
    "Producto A": {
        "cantidad": 10,
        "precio_unitario": 5.0,
        "valor_total":  50.0
    },
    "Producto B": {
        "cantidad": 3,
        "precio_unitario": 12.0,
        "valor_total": 36.0
    },
    "Producto C":{
        "cantidad": 8,
        "precio_unitario": 7.5,
        "valor_total": 60
    }
}

# contar el numoer de productos
print(f"N√∫mero de productos: {len(ventas)}")

# Sumatorio de todas las unidades vendidas
total_unidades = 0
suma_precios_unitarios = 0 
for key, value in ventas.items():
    total_unidades += value["cantidad"]
    suma_precios_unitarios += value["precio_unitario"]

precio_unitario_promedio = round(suma_precios_unitarios / len(ventas), 2)

print(f"Unidades Vendidas: {total_unidades}") 
print(f"Promedio Precio Unitario: {precio_unitario_promedio} ‚Ç¨") 
```

```python
inventario = {
    "manzanas": {"stock": 25, "precio": 0.5, "caducado": False},
    "leche": {"stock": 0, "precio": 1.1, "caducado": True},
    "pan": {"stock": 5, "precio": 0.8, "caducado": False},
    "yogur": {"stock": 2, "precio": 0.9, "caducado": True},
    "huevos": {"stock": 12, "precio": 0.2, "caducado": False},
}

# Mostrar cantidad de productos en el inventario
# Calcular el valor total del inventario (solo productos no caducados y con stock > 0)
# Imprimir una lista de productos caducados (solo los nombres).

# Mostrar cantidad de productos en el inventario

print(f"N√∫mero de productos en el Invent√°rio: {len(inventario)}")

# Calcular el valor total del inventario (solo productos no caducados y con stock > 0)
valor_total = 0
for key, value in inventario.items():
    if value["caducado"] == False and value["stock"] > 0:
        valor_total += value["stock"] * value["precio"]

print(f"Valor total del inventario: {valor_total:.2f}‚Ç¨")

# Imprimir una lista de productos caducados (solo los nombres).

for key, value in inventario.items():
    if value["caducado"]:
        print(f"Producto caducad: {key}")
```

```python
inventario = {
    "manzanas": {"stock": 25, "precio": 0.5, "caducado": False},
    "leche": {"stock": 10, "precio": 1.1, "caducado": False},
    "pan": {"stock": 5, "precio": 0.8, "caducado": False},
    "yogur": {"stock": 7, "precio": 0.9, "caducado": True},
    "huevos": {"stock": 12, "precio": 0.2, "caducado": False},
}

ventas = {
    "manzanas": 10,
    "leche": 5,
    "pan": 3,
    "huevos": 6
}

# 1.	Restar del inventario las unidades vendidas, solo si el producto no est√° caducado.
# 2.	Mostrar el nuevo inventario (producto y stock actualizado).
# 3.	Mostrar el producto con m√°s stock disponible tras las ventas.
# 4.	Mostrar el producto m√°s caro actualmente (por precio unitario).

for key, value in ventas.items():
    if not inventario[key]["caducado"]:
        inventario[key]["stock"] -= value

# Salida 1 & 2
print(f"Listado de Productos")
print(f"-" * 40)
for producto, datos in inventario.items():
    print(f"Producto: {producto}")
    print(f"    Stock: {datos['stock']}")
    print(f"    Precio: {datos['precio']}")
    print(f"    Caducado: {'S√≠' if datos['caducado'] else 'No'}\n")

# 3
mayor_stock = 0
producto_mayor_stock =""
for producto, datos in inventario.items():
    if datos["stock"] > mayor_stock:
        mayor_stock = datos["stock"]
        producto_mayor_stock = producto

print("="*30)
print(f"Producto con mayor Stock")
print(f"\nProducto: {producto_mayor_stock}")
print(f"    Stock: {mayor_stock}")

# 4
mayor_precio = 0
producto_mayor_precio = ""
for producto, datos in inventario.items():
    if datos["precio"] > mayor_precio:
        mayor_precio = datos["precio"]
        producto_mayor_precio = producto

print("="*30)
print(f"Producto con mayor Precio Unit√°rio")
print(f"\nProducto: {producto_mayor_precio}")
print(f"    Precio: {mayor_precio}")
```
