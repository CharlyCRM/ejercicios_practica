```python
# EJERCICIO: CALCULADORA DE NOTAS DE UN CURSO

# bjetivo

# Tienes una lista de estudiantes y sus calificaciones en varias asignaturas. El objetivo es calcular:
# 	1.	La media de cada estudiante.
# 	2.	Qué estudiantes han aprobado (media ≥ 5).
# 	3.	Cuál es la media de la clase.
# 	4.	Quién tiene la nota media más alta.

estudiantes = [
    {"nombre": "Carlos", "notas": [7.5, 8.0, 6.5]},
    {"nombre": "Lucía", "notas": [4.0, 5.0, 3.5]},
    {"nombre": "Miguel", "notas": [9.0, 8.5, 9.5]},
    {"nombre": "Sara", "notas": [6.0, 6.5, 7.0]}
]

# 1.- Media de cada estudiante
print(f"\n--- NOTA MEDIA POR ALUMNO ---")
for estudiante in estudiantes:
    nota_media = sum(estudiante["notas"]) / len(estudiante["notas"])
    print(f"• Estudiante: {estudiante['nombre']}\n      Nota Media: {nota_media:.2f}")

print("-"*40)
# 2.- Estudiante que ha aprobado
print(f"\n--- ESTUDIANTES APROBABADOS ---")
for estudiante in estudiantes:
    nota_media = sum(estudiante["notas"]) / len(estudiante["notas"])
    if nota_media >= 5:
        print(f"• Estudiante: {estudiante['nombre']}\n      Nota Media: {nota_media:.2f}")

# 3.- Media de toda la clase
print(f"\n--- NOTA MEDIA CLASE ---")
suma_medias = 0

for estudiante in estudiantes:
    nota_media = sum(estudiante["notas"]) / len(estudiante["notas"])
    suma_medias += nota_media
media_clase = suma_medias / len(estudiantes)
print(f"          {media_clase}")

# 4.- Nota más alta
print(f"\n--- NOTA MEDIA MÁS ALTA ---")
mejor_estudiante = None
mejor_nota = 0

for estudiante in estudiantes:
    nota_media = sum(estudiante["notas"]) / len(estudiante["notas"])
    if nota_media > mejor_nota:
        mejor_nota = nota_media
        mejor_estudiante = estudiante["nombre"]
print(f"• Estudiante: {mejor_estudiante}\n      Nota Media: {mejor_nota:.2f}")
```

```python
# EJERCICIO: INVENTARIO DE LIBROS

# Tienes una pequeña base de datos de libros en una biblioteca. Cada libro está representado por un diccionario con los siguientes campos:
# 	•	"titulo": título del libro
# 	•	"autor": autor del libro
# 	•	"disponible": True si está disponible para préstamo, False si está prestado

# Debes completar los siguientes pasos:

# 1. Mostrar todos los libros disponibles para préstamo.
# 2. Contar cuántos libros hay por autor.
# 3. Encontrar el primer libro no disponible y mostrar su título y autor.
# 4. Comprobar si existe un libro con el título en el inventario (ignorando mayúsculas/minúsculas).

libros = [
    {"titulo": "Cien años de soledad", "autor": "Gabriel García Márquez", "disponible": True},
    {"titulo": "1984", "autor": "George Orwell", "disponible": False},
    {"titulo": "Crónica de una muerte anunciada", "autor": "Gabriel García Márquez", "disponible": True},
    {"titulo": "El Principito", "autor": "Antoine de Saint-Exupéry", "disponible": False},
    {"titulo": "Fahrenheit 451", "autor": "Ray Bradbury", "disponible": True}
]

# 1.- Mostrar todos los libros disponibles
print(f"--- LIBROS DISPONIBRES PARA PRESTAR ---")
for libro in libros:
    if libro["disponible"]:
        print(f"• Título: {libro['titulo']}\n       Autor: {libro['autor']}")

# 2.- Mostar número libros por autor
print(f"\n--- NÚMERO DE LIBROS POR AUTOR ---")
conteo_por_autor = {}

for libro in libros:
    autor = libro["autor"]
    if autor in conteo_por_autor:
        conteo_por_autor[autor] += 1
    else:
        conteo_por_autor[autor] = 1

for autor, valor in conteo_por_autor.items():
    print(f"• Título: {autor}\n     Número libros: {valor}")

# 3.- Encontrar el primer libro no disponible y mostrar su título y autor.
print(f"\n--- PRIMER LIBRO NO DISPONIBLE ---")

for libro in libros:
    if libro["disponible"] == False:
        print(f"• Título: {libro['titulo']}\n   Autor: {libro['autor']}\n   Disponibilidad {libro['disponible']}")
        break

# 4.- Comprobar si existe un libro con el título en el inventario (ignorando mayúsculas/minúsculas).
print(f"\n--- COMPROBAR LIBRO ---")
encontrado = False

libro_buscar = input(f"Introduzca el nombre del libro a buscar: ").lower()

for libro in libros:
    if libro_buscar in libro["titulo"].lower(): # in: permite realizar busquedar parciales
        encontrado = True                       # ==: buscar coincidencia exacta
        break

if encontrado:
    print("Libro encontrado")
else:
    print("Libro no encontrado")
```

```python
# EJERCICIO: GESTOR DE COMPRAS EN SUPERMERCADO

# Objetivo

# Simular una lista de productos comprados por un cliente. Cada producto tiene nombre, precio y cantidad. 
# Se deben realizar operaciones sobre esta lista mediante funciones.

# Tareas (usa una función por cada punto):
# 	1.	mostrar_carrito(lista): muestra todos los productos con nombre, precio unitario, cantidad y subtotal (precio × cantidad).
# 	2.	calcular_total(lista): devuelve el importe total de la compra.
# 	3.	producto_mas_caro(lista): devuelve el producto con mayor precio unitario.
# 	4.	añadir_producto(lista, nombre, precio, cantidad): añade un nuevo producto al carrito.

carrito = [
    {"nombre": "manzanas", "precio": 0.5, "cantidad": 4},
    {"nombre": "leche", "precio": 1.2, "cantidad": 2},
    {"nombre": "pan", "precio": 1.0, "cantidad": 1}
]

# 1.- mostrar_carrito(lista): muestra todos los productos con nombre, precio unitario, cantidad y subtotal (precio × cantidad).
def mostrar_carrito(lista:list) -> None:
    "Muestra todos los productos incluidos en el carrito"

    print("\n🛒 Carrito de la compra:\n")
    for producto in lista:
        subtotal = producto["precio"] * producto["cantidad"]
        print(f"- {producto['nombre'].capitalize()}")
        print(f"  Precio unitario: €{producto['precio']:.2f}")
        print(f"  Cantidad: {producto['cantidad']}")
        print(f"  Subtotal: €{subtotal:.2f}\n")

# 2.- calcular_total(lista): devuelve el importe total de la compra.
def calcular_total(lista:list)-> float:
    "Devuelve el importe total de cada compra"

    importe_total = 0

    for producto in lista:
        subtotal = producto["precio"] * producto["cantidad"]
        importe_total += subtotal
    return importe_total

#3.- producto_mas_caro(lista): devuelve el producto con mayor precio unitario.
def producto_mas_caro(lista:list) -> tuple:
    "Devuelve el producto más caro"

    nombre_producto = None
    mayor_precio = 0

    for producto in lista:
        if producto["precio"] > mayor_precio:
            mayor_precio = producto["precio"]
            nombre_producto = producto["nombre"]
    return (nombre_producto, mayor_precio)

# 4.- añadir_producto(lista, nombre, precio, cantidad): añade un nuevo producto al carrito.
def añadir_producto(lista:list, nombre:str, precio:float = 0.0, cantidad:float = 0.0)-> None:
    "Añade un nuevo producto a la lista"
    lista.append({
            "nombre": nombre,
            "precio": precio,
            "cantidad": cantidad
        })
    print(f"🛒 Producto '{nombre}' añadido al carrito.")

        
if __name__ == "__main__":

    mostrar_carrito(carrito)
    print(f"💵 Importe total del carrito: {calcular_total(carrito)} €")
    nombre, precio = producto_mas_caro(carrito)
    print(f"🛒 Producto más caro: {nombre.capitalize()} ({precio} €)")
    añadir_producto(carrito, "peras", 0.4, 2)
```

```python
# EJERCICIO: SISTEMA DE GESTIÓN DE PRODUCTOS CON POO

# Objetivo: Crear una clase Producto que represente un producto de un carrito, y otra clase Carrito que administre una lista de productos.

# Requisitos:
# 	1.	Clase Producto:
# 	•	Atributos: nombre, precio y cantidad.
# 	•	Método: subtotal() → devuelve el precio × cantidad.

# 	2.	Clase Carrito:
# 	•	Atributo: una lista de productos.
# 	•	Métodos:
# 	•	añadir_producto(producto: Producto)
# 	•	mostrar_productos() → similar a mostrar_carrito() pero usando objetos.
# 	•	calcular_total() → suma de todos los subtotales.
# 	•	producto_mas_caro() → devuelve el producto con mayor precio unitario.

class Producto:
    def __init__(self, nombre: str, precio: float = 0.0, cantidad: float = 0.0):
        self.nombre = nombre
        self.precio = precio
        self.cantidad = cantidad

    def subtotal(self):
        return self.precio * self.cantidad


class Carrito:
    def __init__(self, lista_productos: list = None):
        self.lista_productos = lista_productos if lista_productos is not None else []

    def añadir_producto(self, producto) -> None:
        self.lista_productos.append(producto)
        print(f"🛒 Producto '{producto.nombre}' agregado correctamente")

    def mostrar_productos(self) -> None:
        print("\n--- 🛒 CARRITO DE LA COMPRA ---\n")
        for producto in self.lista_productos:
            print(f"- {producto.nombre.capitalize()}")
            print(f"- Precio unitario €{producto.precio:.2f}")
            print(f"- Cantidad: {producto.cantidad}")
            print(f"- Subtotal: {producto.subtotal():.2f}")
            print("-" * 30)

    def calcular_total(self) -> float:
        return sum(producto.subtotal() for producto in self.lista_productos)

    def producto_mas_caro(self) -> tuple:
        if not self.lista_productos:
            return (None, 0)
        producto_mayor = max(self.lista_productos, key=lambda p: p.precio)
        return (producto_mayor.nombre, producto_mayor.precio)

if __name__ == "__main__":
    # Crear instancia del carrito
    carrito = Carrito()

    # Crear productos
    manzana = Producto("manzana", 0.5, 4)
    leche = Producto("leche", 1.2, 2)
    pan = Producto("pan", 1.0, 1)

    # Añadir productos al carrito
    carrito.añadir_producto(manzana)
    carrito.añadir_producto(leche)
    carrito.añadir_producto(pan)

    # Mostrar el contenido del carrito
    carrito.mostrar_productos()

    # Mostrar el total
    print(f"\n💰 Total a pagar: {carrito.calcular_total():.2f} €")

    # Mostrar el producto más caro
    nombre, precio = carrito.producto_mas_caro()
    print(f"🔝 Producto más caro: {nombre.capitalize()} ({precio:.2f} €)")
```

```python
# EJERCICIO: “BATALLA DE POKÉMON” CON CLASES
# Objetivo:

# Crear una simulación básica de combate entre dos Pokémon usando Programación Orientada a Objetos. 

# Requisitos del ejercicio

# 1. Clase Pokemon
# 	•	Atributos:
# 	•	nombre (str)
# 	•	tipo (str)
# 	•	vida (int)
# 	•	ataque (int)
# 	•	Métodos:
# 	•	atacar(objetivo: Pokemon) → reduce la vida del oponente en función del ataque.
# 	•	esta_vivo() → devuelve True si la vida > 0.
# 	•	__str__() → para imprimir el estado del Pokémon.

# 2. Clase BatallaPokemon
# 	•	Atributos:
# 	•	pokemon1 (instancia de Pokemon)
# 	•	pokemon2 (instancia de Pokemon)
# 	•	Métodos:
# 	•	iniciar_batalla() → turnos alternos de ataque hasta que uno quede con 0 de vida. Muestra cada paso por consola.

# Permite usar anotaciones de tipo con clases que aún no están definidas completamente.
# Útil para referirse a la propia clase dentro de su definición sin necesidad de comillas.
# En este scrip lo uso por ejemplo dentro del método atacar en "objetivo: Pokemon"
from __future__ import annotations

class Pokemon:
    def __init__(self, nombre:str, tipo:str, vida:int, ataque:int):
        self.nombre = nombre
        self.tipo = tipo
        self.vida = vida
        self.ataque = ataque

    def atacar(self, objetivo: Pokemon) -> None:
        "Reduce la vida del oponente en función del ataque"
        vida_original = objetivo.vida
        nueva_vida = vida_original - self.ataque
        objetivo.vida = nueva_vida
        print(f"💥 {self.nombre} ataca a {objetivo.nombre} y causa {self.ataque} puntos de daño")

    def esta_vivo(self) -> bool:
        return self.vida > 0
        
    def __str__(self) -> str:
        return (
            f"🐾 Nombre: {self.nombre.capitalize()}\n"
            f"🔋 Tipo: {self.tipo.capitalize()}\n"
            f"❤️ Vida: {self.vida}\n"
            f"💥 Ataque: {self.ataque}"
        )

class BatallaPokemon:
    def __init__(self, pokemon1: Pokemon, pokemon2: Pokemon):
        self.pokemon1 = pokemon1
        self.pokemon2 = pokemon2

    def iniciar_batalla(self):
        "Turnos alternos de ataque hasta que uno quede con 0 vida"

        print(f"\n--- INICIO BATALLA POKÉMON ----")
        print(f"     {self.pokemon1.nombre} VS {self.pokemon2.nombre}")

        contador_rounds = 1
        while self.pokemon1.esta_vivo() and self.pokemon2.esta_vivo():
            print(f"\n--- ROUND {contador_rounds} ---")

            # Alternar turnos según si el número de ronda es impar o par
            if contador_rounds % 2 != 0:
                atacante, defensor = self.pokemon1, self.pokemon2
            else:
                atacante, defensor = self.pokemon2, self.pokemon1

            atacante.atacar(defensor)

            # Solo contraataca si sigue con vida
            if defensor.esta_vivo():
                defensor.atacar(atacante)

            # Mostrar estado de ambos Pokémon tras el round
            print(f"\n{self.pokemon1}")
            print(f"\n{self.pokemon2}")

            contador_rounds += 1

        print("\n--- FIN DE LA BATALLA ---")
        if self.pokemon1.esta_vivo():
            print(f"🏆 ¡{self.pokemon1.nombre} gana la batalla!")
        else:
            print(f"🏆 ¡{self.pokemon2.nombre} gana la batalla!")



if __name__ == "__main__":
    # Creacion de objeto pokeon
    charizard = Pokemon("Charizard","fuego",30, 5)
    pikachu = Pokemon("Pikachu","electrico",10, 6)
    bulbasur = Pokemon("Bulbasur", "planta", 25, 5)

    combate = BatallaPokemon(charizard, pikachu)
    combate.iniciar_batalla()

```
