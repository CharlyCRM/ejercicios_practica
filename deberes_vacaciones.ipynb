```python
inventario = {
    "manzanas": {"stock": 25, "precio": 0.5, "caducado": False},
    "leche": {"stock": 10, "precio": 1.1, "caducado": False},
    "pan": {"stock": 5, "precio": 0.8, "caducado": False},
    "yogur": {"stock": 7, "precio": 0.9, "caducado": True},
    "huevos": {"stock": 12, "precio": 0.2, "caducado": False},
}

ventas = {
    "manzanas": 10,
    "leche": 5,
    "pan": 3,
    "huevos": 6
}

# 1.	Restar del inventario las unidades vendidas, solo si el producto no est√° caducado.
# 2.	Mostrar el nuevo inventario (producto y stock actualizado).
# 3.	Mostrar el producto con m√°s stock disponible tras las ventas.
# 4.	Mostrar el producto m√°s caro actualmente (por precio unitario).

for key, value in ventas.items():
    if not inventario[key]["caducado"]:
        inventario[key]["stock"] -= value

# Salida 1 & 2
print(f"Listado de Productos")
print(f"-" * 40)
for producto, datos in inventario.items():
    print(f"Producto: {producto}")
    print(f"    Stock: {datos['stock']}")
    print(f"    Precio: {datos['precio']}")
    print(f"    Caducado: {'S√≠' if datos['caducado'] else 'No'}\n")

# 3
mayor_stock = 0
producto_mayor_stock =""
for producto, datos in inventario.items():
    if datos["stock"] > mayor_stock:
        mayor_stock = datos["stock"]
        producto_mayor_stock = producto

print("="*30)
print(f"Producto con mayor Stock")
print(f"\nProducto: {producto_mayor_stock}")
print(f"    Stock: {mayor_stock}")

# 4
mayor_precio = 0
producto_mayor_precio = ""
for producto, datos in inventario.items():
    if datos["precio"] > mayor_precio:
        mayor_precio = datos["precio"]
        producto_mayor_precio = producto

print("="*30)
print(f"Producto con mayor Precio Unit√°rio")
print(f"\nProducto: {producto_mayor_precio}")
print(f"    Precio: {mayor_precio}")
```

```python
# EJERCICIO: CALCULADORA DE NOTAS DE UN CURSO

# bjetivo

# Tienes una lista de estudiantes y sus calificaciones en varias asignaturas. El objetivo es calcular:
# 	1.	La media de cada estudiante.
# 	2.	Qu√© estudiantes han aprobado (media ‚â• 5).
# 	3.	Cu√°l es la media de la clase.
# 	4.	Qui√©n tiene la nota media m√°s alta.

estudiantes = [
    {"nombre": "Carlos", "notas": [7.5, 8.0, 6.5]},
    {"nombre": "Luc√≠a", "notas": [4.0, 5.0, 3.5]},
    {"nombre": "Miguel", "notas": [9.0, 8.5, 9.5]},
    {"nombre": "Sara", "notas": [6.0, 6.5, 7.0]}
]

# 1.- Media de cada estudiante
print(f"\n--- NOTA MEDIA POR ALUMNO ---")
for estudiante in estudiantes:
    nota_media = sum(estudiante["notas"]) / len(estudiante["notas"])
    print(f"‚Ä¢ Estudiante: {estudiante['nombre']}\n      Nota Media: {nota_media:.2f}")

print("-"*40)
# 2.- Estudiante que ha aprobado
print(f"\n--- ESTUDIANTES APROBABADOS ---")
for estudiante in estudiantes:
    nota_media = sum(estudiante["notas"]) / len(estudiante["notas"])
    if nota_media >= 5:
        print(f"‚Ä¢ Estudiante: {estudiante['nombre']}\n      Nota Media: {nota_media:.2f}")

# 3.- Media de toda la clase
print(f"\n--- NOTA MEDIA CLASE ---")
suma_medias = 0

for estudiante in estudiantes:
    nota_media = sum(estudiante["notas"]) / len(estudiante["notas"])
    suma_medias += nota_media
media_clase = suma_medias / len(estudiantes)
print(f"          {media_clase}")

# 4.- Nota m√°s alta
print(f"\n--- NOTA MEDIA M√ÅS ALTA ---")
mejor_estudiante = None
mejor_nota = 0

for estudiante in estudiantes:
    nota_media = sum(estudiante["notas"]) / len(estudiante["notas"])
    if nota_media > mejor_nota:
        mejor_nota = nota_media
        mejor_estudiante = estudiante["nombre"]
print(f"‚Ä¢ Estudiante: {mejor_estudiante}\n      Nota Media: {mejor_nota:.2f}")
```

```python
# EJERCICIO: INVENTARIO DE LIBROS

# Tienes una peque√±a base de datos de libros en una biblioteca. Cada libro est√° representado por un diccionario con los siguientes campos:
# 	‚Ä¢	"titulo": t√≠tulo del libro
# 	‚Ä¢	"autor": autor del libro
# 	‚Ä¢	"disponible": True si est√° disponible para pr√©stamo, False si est√° prestado

# Debes completar los siguientes pasos:

# 1. Mostrar todos los libros disponibles para pr√©stamo.
# 2. Contar cu√°ntos libros hay por autor.
# 3. Encontrar el primer libro no disponible y mostrar su t√≠tulo y autor.
# 4. Comprobar si existe un libro con el t√≠tulo en el inventario (ignorando may√∫sculas/min√∫sculas).

libros = [
    {"titulo": "Cien a√±os de soledad", "autor": "Gabriel Garc√≠a M√°rquez", "disponible": True},
    {"titulo": "1984", "autor": "George Orwell", "disponible": False},
    {"titulo": "Cr√≥nica de una muerte anunciada", "autor": "Gabriel Garc√≠a M√°rquez", "disponible": True},
    {"titulo": "El Principito", "autor": "Antoine de Saint-Exup√©ry", "disponible": False},
    {"titulo": "Fahrenheit 451", "autor": "Ray Bradbury", "disponible": True}
]

# 1.- Mostrar todos los libros disponibles
print(f"--- LIBROS DISPONIBRES PARA PRESTAR ---")
for libro in libros:
    if libro["disponible"]:
        print(f"‚Ä¢ T√≠tulo: {libro['titulo']}\n       Autor: {libro['autor']}")

# 2.- Mostar n√∫mero libros por autor
print(f"\n--- N√öMERO DE LIBROS POR AUTOR ---")
conteo_por_autor = {}

for libro in libros:
    autor = libro["autor"]
    if autor in conteo_por_autor:
        conteo_por_autor[autor] += 1
    else:
        conteo_por_autor[autor] = 1

for autor, valor in conteo_por_autor.items():
    print(f"‚Ä¢ T√≠tulo: {autor}\n     N√∫mero libros: {valor}")

# 3.- Encontrar el primer libro no disponible y mostrar su t√≠tulo y autor.
print(f"\n--- PRIMER LIBRO NO DISPONIBLE ---")

for libro in libros:
    if libro["disponible"] == False:
        print(f"‚Ä¢ T√≠tulo: {libro['titulo']}\n   Autor: {libro['autor']}\n   Disponibilidad {libro['disponible']}")
        break

# 4.- Comprobar si existe un libro con el t√≠tulo en el inventario (ignorando may√∫sculas/min√∫sculas).
print(f"\n--- COMPROBAR LIBRO ---")
encontrado = False

libro_buscar = input(f"Introduzca el nombre del libro a buscar: ").lower()

for libro in libros:
    if libro_buscar in libro["titulo"].lower(): # in: permite realizar busquedar parciales
        encontrado = True                       # ==: buscar coincidencia exacta
        break

if encontrado:
    print("Libro encontrado")
else:
    print("Libro no encontrado")
```

```python
# EJERCICIO: GESTOR DE COMPRAS EN SUPERMERCADO

# Objetivo

# Simular una lista de productos comprados por un cliente. Cada producto tiene nombre, precio y cantidad. 
# Se deben realizar operaciones sobre esta lista mediante funciones.

# Tareas (usa una funci√≥n por cada punto):
# 	1.	mostrar_carrito(lista): muestra todos los productos con nombre, precio unitario, cantidad y subtotal (precio √ó cantidad).
# 	2.	calcular_total(lista): devuelve el importe total de la compra.
# 	3.	producto_mas_caro(lista): devuelve el producto con mayor precio unitario.
# 	4.	a√±adir_producto(lista, nombre, precio, cantidad): a√±ade un nuevo producto al carrito.

carrito = [
    {"nombre": "manzanas", "precio": 0.5, "cantidad": 4},
    {"nombre": "leche", "precio": 1.2, "cantidad": 2},
    {"nombre": "pan", "precio": 1.0, "cantidad": 1}
]

# 1.- mostrar_carrito(lista): muestra todos los productos con nombre, precio unitario, cantidad y subtotal (precio √ó cantidad).
def mostrar_carrito(lista:list) -> None:
    "Muestra todos los productos incluidos en el carrito"

    print("\nüõí Carrito de la compra:\n")
    for producto in lista:
        subtotal = producto["precio"] * producto["cantidad"]
        print(f"- {producto['nombre'].capitalize()}")
        print(f"  Precio unitario: ‚Ç¨{producto['precio']:.2f}")
        print(f"  Cantidad: {producto['cantidad']}")
        print(f"  Subtotal: ‚Ç¨{subtotal:.2f}\n")

# 2.- calcular_total(lista): devuelve el importe total de la compra.
def calcular_total(lista:list)-> float:
    "Devuelve el importe total de cada compra"

    importe_total = 0

    for producto in lista:
        subtotal = producto["precio"] * producto["cantidad"]
        importe_total += subtotal
    return importe_total

#3.- producto_mas_caro(lista): devuelve el producto con mayor precio unitario.
def producto_mas_caro(lista:list) -> tuple:
    "Devuelve el producto m√°s caro"

    nombre_producto = None
    mayor_precio = 0

    for producto in lista:
        if producto["precio"] > mayor_precio:
            mayor_precio = producto["precio"]
            nombre_producto = producto["nombre"]
    return (nombre_producto, mayor_precio)

# 4.- a√±adir_producto(lista, nombre, precio, cantidad): a√±ade un nuevo producto al carrito.
def a√±adir_producto(lista:list, nombre:str, precio:float = 0.0, cantidad:float = 0.0)-> None:
    "A√±ade un nuevo producto a la lista"
    lista.append({
            "nombre": nombre,
            "precio": precio,
            "cantidad": cantidad
        })
    print(f"üõí Producto '{nombre}' a√±adido al carrito.")

        
if __name__ == "__main__":

    mostrar_carrito(carrito)
    print(f"üíµ Importe total del carrito: {calcular_total(carrito)} ‚Ç¨")
    nombre, precio = producto_mas_caro(carrito)
    print(f"üõí Producto m√°s caro: {nombre.capitalize()} ({precio} ‚Ç¨)")
    a√±adir_producto(carrito, "peras", 0.4, 2)
```

```python
# EJERCICIO: SISTEMA DE GESTI√ìN DE PRODUCTOS CON POO

# Objetivo: Crear una clase Producto que represente un producto de un carrito, y otra clase Carrito que administre una lista de productos.

# Requisitos:
# 	1.	Clase Producto:
# 	‚Ä¢	Atributos: nombre, precio y cantidad.
# 	‚Ä¢	M√©todo: subtotal() ‚Üí devuelve el precio √ó cantidad.

# 	2.	Clase Carrito:
# 	‚Ä¢	Atributo: una lista de productos.
# 	‚Ä¢	M√©todos:
# 	‚Ä¢	a√±adir_producto(producto: Producto)
# 	‚Ä¢	mostrar_productos() ‚Üí similar a mostrar_carrito() pero usando objetos.
# 	‚Ä¢	calcular_total() ‚Üí suma de todos los subtotales.
# 	‚Ä¢	producto_mas_caro() ‚Üí devuelve el producto con mayor precio unitario.

class Producto:
    def __init__(self, nombre: str, precio: float = 0.0, cantidad: float = 0.0):
        self.nombre = nombre
        self.precio = precio
        self.cantidad = cantidad

    def subtotal(self):
        return self.precio * self.cantidad


class Carrito:
    def __init__(self, lista_productos: list = None):
        self.lista_productos = lista_productos if lista_productos is not None else []

    def a√±adir_producto(self, producto) -> None:
        self.lista_productos.append(producto)
        print(f"üõí Producto '{producto.nombre}' agregado correctamente")

    def mostrar_productos(self) -> None:
        print("\n--- üõí CARRITO DE LA COMPRA ---\n")
        for producto in self.lista_productos:
            print(f"- {producto.nombre.capitalize()}")
            print(f"- Precio unitario ‚Ç¨{producto.precio:.2f}")
            print(f"- Cantidad: {producto.cantidad}")
            print(f"- Subtotal: {producto.subtotal():.2f}")
            print("-" * 30)

    def calcular_total(self) -> float:
        return sum(producto.subtotal() for producto in self.lista_productos)

    def producto_mas_caro(self) -> tuple:
        if not self.lista_productos:
            return (None, 0)
        producto_mayor = max(self.lista_productos, key=lambda p: p.precio)
        return (producto_mayor.nombre, producto_mayor.precio)

if __name__ == "__main__":
    # Crear instancia del carrito
    carrito = Carrito()

    # Crear productos
    manzana = Producto("manzana", 0.5, 4)
    leche = Producto("leche", 1.2, 2)
    pan = Producto("pan", 1.0, 1)

    # A√±adir productos al carrito
    carrito.a√±adir_producto(manzana)
    carrito.a√±adir_producto(leche)
    carrito.a√±adir_producto(pan)

    # Mostrar el contenido del carrito
    carrito.mostrar_productos()

    # Mostrar el total
    print(f"\nüí∞ Total a pagar: {carrito.calcular_total():.2f} ‚Ç¨")

    # Mostrar el producto m√°s caro
    nombre, precio = carrito.producto_mas_caro()
    print(f"üîù Producto m√°s caro: {nombre.capitalize()} ({precio:.2f} ‚Ç¨)")
```
